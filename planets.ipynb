```python
%matplotlib inline
from pylab import *
from IPython.display import HTML
```

```python
#imports
import fuzzy
from Levenshtein import distance
```

```python
#set up the phonetic tools
soundex = fuzzy.Soundex(4)
dmeta = fuzzy.DMetaphone()
```

```python
#get Data
with open('data/planets.txt', 'r') as PLANETS:
    languages = dict()
    for line in PLANETS:
        languages[line.strip().split('\t')[0]] = line.strip().split('\t')[1:]
```

```python
    
    #generate the Levenshtein of the D-Meta for each planet from for each Language from English
    coords = dict()
    for language in languages.keys()[1:]:
        distances = list()
        for i, english in enumerate(languages['English']):
            try:
                distances.append(distance(english, dmeta(languages[language][i])[0]))
            except Exception:
                pass
        coords[language] = distances
    
    for coord in coords.values():
        plot(coord)
    
```

```python
    #tabulate the data in html
    tablestring = '<table>'
    for i, name in enumerate(languages['English']):
        tablestring += ('<tr><td colspan="6"><h2>%s</h2></td></tr>' % name)
        for language in languages.keys()[1:]:
            try:
                tablestring += "<tr><td>%s</td><td>SOUNDEX</td><td>%s</td><td>%s</td><td>%s</td><td>%s</td></tr>" % (language, languages['English'][i], languages[language][i], soundex(languages['English'][i]), soundex(languages[language][i]))
                tablestring += "<tr><td>%s</td><td>NYSIIS</td><td>%s</td><td>%s</td><td>%s</td><td>%s</td></tr>" % (language, languages['English'][i], languages[language][i], fuzzy.nysiis(languages['English'][i]), fuzzy.nysiis(languages[language][i]))
                tablestring += "<tr><td>%s</td><td>D-meta</td><td>%s</td><td>%s</td><td>%s</td><td>%s</td></tr>" % (language, languages['English'][i], languages[language][i], dmeta(languages['English'][i])[0], dmeta(languages[language][i])[0])
            except IndexError:
                tablestring += '<tr><td colspan="2">%s</td><td>%s</td><td colspan="3"><em>NO ENTRY</em></td></tr>' % (language, languages['English'][i])

HTML(tablestring+'</table>')
```
